```python
import pandas as pd
import json
import geopandas as gpd
```

```python
county = gpd.read_file('county.json')
```

```python
county.head()
```

```python
maindata = pd.read_csv('data/maindata.csv')
```

```python
maindata.head()
```

```python
maindata.STATEFIP = pd.Series(maindata.STATEFIP, dtype = 'int32')
maindata.COUNTYFIPS = pd.Series(maindata.COUNTYFIPS, dtype = 'int32')
```

```python
county.STATEFP = pd.Series(county.STATEFP, dtype = 'float64')
county.COUNTYFP = pd.Series(county.COUNTYFP, dtype = 'float64')
```

```python
county.dtypes
```

```python
newdata = pd.merge(county, maindata, how = 'left', left_on = ['STATEFP', 'COUNTYFP'], right_on=['STATEFIP','COUNTYFIPS'])
```

```python
newdata.dtypes
```

```python
newdata.to_file('countymerged.json', driver = 'GeoJSON')
```
